Project Summary -

Paisabazaar Banking Fraud / Credit Risk Analysis

This project aims to analyze Paisabazaar banking data to identify high-risk and potentially fraudulent customers using their financial and credit behavior. The dataset consists of 100,000 records and 28 variables, including customer demographics, income details, credit usage, repayment behavior, and credit score. Since the dataset does not contain a direct fraud label, customers with a “Poor” credit score were considered high-risk, which is a common and practical assumption in banking and fintech industries.

The first step of the project involved data loading and basic cleaning. Rows with missing credit score values were removed, and data types were verified to ensure consistency. Important features related to credit risk were selected, such as annual income, number of delayed payments, outstanding debt, credit utilization ratio, and monthly balance. Extreme values were carefully reviewed but not aggressively removed, as such values often carry meaningful information in fraud and risk analysis.

Next, Exploratory Data Analysis (EDA) was performed to understand customer behavior patterns. Bar charts and box plots were used to compare different features across credit score categories (Good, Standard, and Poor). This visual analysis helped in identifying clear differences between low-risk and high-risk customers.

The EDA revealed several important insights. Customers with poor credit scores had significantly higher delayed payments, which emerged as the strongest indicator of risk. These customers also showed higher outstanding debt and higher credit utilization, indicating over-dependence on credit. In contrast, customers with good credit scores maintained lower debt levels, paid on time, and used credit more responsibly. Income also played a key role, as high-risk customers generally had lower annual income, making repayment more difficult. Additionally, younger customers were more likely to fall into the high-risk category, possibly due to shorter credit history and unstable income.

In conclusion, the project demonstrates that credit behavior variables can effectively identify risky or fraudulent customers even in the absence of a direct fraud label. The insights from this analysis can help fintech platforms like Paisabazaar improve loan approval decisions, reduce default and fraud losses, and implement early risk detection systems.

## Problem Statement

Fintech platforms like Paisabazaar handle a large number of customer loan and credit applications every day. Due to the high volume of transactions, it becomes difficult to manually identify high-risk or potentially fraudulent customers. Incorrect credit decisions can lead to loan defaults, financial losses, and reduced trust in the platform.

The challenge addressed in this project is to analyze customer credit and financial behavior to identify risky customers at an early stage. Since the dataset does not contain a direct fraud indicator, the problem is approached by using credit score as a proxy for risk, where customers with a “Poor” credit score are treated as high-risk or fraud-prone.

The objective of this project is to perform data cleaning, exploratory data analysis (EDA), and behavioral analysis on Paisabazaar banking data to uncover patterns related to delayed payments, debt levels, income, and credit utilization. By identifying the key factors that differentiate low-risk and high-risk customers, the project aims to support better credit decision-making, early risk detection, and fraud prevention in a fintech environment.

##Business Objective

The primary business objective of this project is to identify high-risk and potentially fraudulent customers at an early stage using their credit and financial behavior. By analyzing Paisabazaar banking data, the goal is to understand the patterns that lead to poor credit performance and increased default risk.

This analysis aims to help the business reduce financial losses caused by loan defaults and fraud, improve the quality of credit approvals, and ensure that genuine customers receive faster and more accurate decisions. By identifying key risk indicators such as delayed payments, high outstanding debt, low income, and high credit utilization, the business can apply risk-based rules and policies while approving loans or credit cards.

Overall, the objective is to support data-driven decision-making that enhances customer trust, minimizes credit risk, and improves the long-term profitability and stability of the fintech platform.
